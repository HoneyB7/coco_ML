- 주성분이란?
    - 데이터들의 <font color = 'blue'><b>분산이 가장 큰 방향벡터</b></font>를 의미함.<br><br>

- 주성분 분석이 이용되는 예
    - 영상 인식
    - 통계 데이터 분석 (주성분 찾기)
    - 데이터 압축 (차원 감소)
    - 노이즈 제거 <br><br>

- 주성분 분석의 장/단점
    - 장점 : 데이터의 복잡도를 줄이고, 가장 중요한 속성을 구별한다.
    - 단점 : 유용한 정보를 날려버릴 수도 있다.
    - 활용 : 수치형 값 <br><br>
    
- 주성분 분석을 통해 차원 축소 하는 방법
    - 데이터를 불러온다.
    - 평균으로 데이터를 차감시켜준다.
    - 공분산 행렬 계산.
    - 고유값을 계산하고, 공분산 행렬의 고유값도 계산.
    - 고유값을 가장 큰 값부터 가장 작은 값까지 정렬
    - 큰 고유값부터 N개의 고유 벡터를 사용.
    - N개의 고유 값으로 생성된 새로운 공간으로 변환. <br><br>
    
- 공분산 행렬이 의미하는 것
    - 성분간의 상관관계를 나타내는 행렬
    - 성분간의 <b>상관관계가 최소</b>가 되는 주성분을 찾기 위해 공분산 행렬을 계산하는 것! (즉, 성분간의 독립성이 최대가 되도록 하는 성분을 찾기 위함)<br><br>
    
- 고유값과 고유벡터란?
    - 고유벡터
        - 행렬 A를 선형변환으로 봤을 때, <b> 선형 변환 A에 의한 변환 결과가 자기 자신의 상수배가 되는 0이 아닌 벡터<b><br>
    - 고유값
        - 위의 상수배 값 <br>

```python
from numpy import *

def loadDataSet(fileName, delim='\t'):
    fr = open(fileName)
    stringArr = [line.strip().split(delim) for line in fr.readlines()]
    datArr = [map(float,line) for line in stringArr]
    return mat(datArr)

def ReduceDImention(eigVals, eigVects, meanRemoved, topNfeat = 100):
    eigValInd = argsort(eigVals)              # sort, sort goes smallest to largest
    eigValInd = eigValInd[:-(topNfeat+1):-1]  # cut off unwanted dimensions
    redEigVects = eigVects[:,eigValInd]       # reorganize eig vects largest to smallest
    lowDDataMat = meanRemoved * redEigVects   # transform data into new dimensions
    reconMat = (lowDDataMat * redEigVects.T) + meanVals
    return lowDDataMat, reconMat

def MeanCov(datMat):
    meanVals = mean(dataMat, axis=0)
    meanRemoved = dataMat - meanVals  # remove mean
    covMat = cov(meanRemoved, rowvar=0)
    
    return covMat

def replaceNanWithMean(datMat): 
    numFeat = shape(datMat)[1]
    
    for i in range(numFeat):
        meanVal = mean(datMat[nonzero(~isnan(datMat[:,i].A))[0],i]) 
        datMat[nonzero(isnan(datMat[:,i].A))[0],i] = meanVal  
    return datMat

def Plot(dataMat, lowDMat, reconMat):
    from mpl_toolkits.mplot3d import Axes3D
    import matplotlib.pyplot as plt

    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    x1 = []; y1 = []; z1 = [];
    x2 = []; y2 = []; z2 = [];
    
    ax.scatter(dataMat[:,0].flatten().A[0], dataMat[:,1].flatten().A[0], c='b', marker='o')
    ax.scatter(reconMat[:,0].flatten().A[0], reconMat[:,1].flatten().A[0], c='r', marker='*')

    ax.set_title('dataMat') 
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')

    plt.show()
```

```python
dataMat = loadDataSet('character.txt')   # 데이터 불러옴
newMat = replaceNanWithMean(dataMat)     # NaN값 평균값으로 바꿈

meanVals = mean(dataMat, axis=0)         # 평균값 계산
meanRemoved = dataMat - meanVals         # 평균값 빼줌, 원점으로 이동
    
covMat = MeanCov(meanRemoved)            # 공분산 구하기

eigValue, eigVector = linalg.eig(mat(covMat))   # 고유값 고유벡터 구하기
lowDMat, reconMat = ReduceDImention(eigValue, eigVector, meanRemoved, 1)   # lowD : 차원 축소 벡터, reconMat : 차원이 축소된 Mat

Plot(dataMat, lowDMat, reconMat)
```
